# Hipótese e Mecanismo:

### Como a organização visa aumentar a cooperação economica e motivar os países membros a competirem entre sí por investimentos de mercado e industrialização, é possivel assumir que os niveis de desemprego diminuem por necessidade de mais mão de obra mesmo com o incremento da técnologia. E como os investimentos e a produção aumentam, o PIB per capta tende a aumentar tambem.




### primeiramente é importante definir esta organização e quais são seus integrantes antes e prosseguirmos para a pesquisa de dados:

    A UNIDO, criada em 1966 e posteriormente convertida em agencia especializada no ano de 1985, foi criada com a espectativa de acelerar o Desenvolvimento Industrial Sustentavel e Inclusivo (I SID). Para isso a agencia leva em consideração alguns métodos inclusivos, como as idéias de Prosperidade Compartilhada, Desenvolvimento da competitividade e a Promoção da qualidade de vida.
    Contendo atualmente 171 estados-membro a ONU organiza a atuação de cada país com o intuito de oferecer mutuamente uma Cooperação técnica, Serviços analiticos e de politicas industriais, Serviços de implemtentação de normas de conformidade industrial e, finalmente, o Estabelecimente de parcerias entre eles.   




### É importante ressaltar que como são 171 países na organização, eu utilizarei apenas os países emergentes; já que são estes países que teoricamente seriam mais afetados pela entrada na UNIDO em questões de superação industrial e laboral.
